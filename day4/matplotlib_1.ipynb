```python
import matplotlib.pyplot as plt
```

```python
x = range(1, 8)
#在1到20之间随机生成7个数据
y = [11, 23, 15, 18, 20, 12, 16]
plt.plot(x, y, color='red', alpha=0.8, marker='o', linestyle='--', linewidth=3)
plt.show()
```

```python
#设置图片的大小和保存
import random

x = range(2, 26, 2)
y = [random.randint(15, 30) for i in x]
#设置画布
plt.figure(figsize=(20, 8), dpi=80)
x_ticks_label = [f"{i}:00" for i in x]
plt.xticks(x, x_ticks_label, rotation=45)
y_ticks_label = [f'{i}℃' for i in range(min(y), max(y) + 1)]
plt.yticks(range(min(y), max(y) + 1), y_ticks_label)
plt.plot(x, y)
plt.show()
# plt.savefig('test.png')#保存图片
# plt.savefig('test.svg')#保存矢量图
```

```python
#显示中文
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [20, 25, 28, 22, 26, 24, 21, 23, 27]
from matplotlib import font_manager

my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc')
plt.xlabel('时间', fontproperties=my_font)
plt.ylabel('温度', fontproperties=my_font)
plt.title('温度变化图', fontproperties=my_font)
ply = plt.plot(x, y)
plt.show()
```

```python
#一图多线
from matplotlib import font_manager

my_font = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc', size=18)
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [20, 25, 28, 22, 26, 24, 21, 23, 27]
y2 = [18, 19, 22, 25, 28, 26, 23, 24, 27]
y3 = [15, 16, 19, 22, 25, 23, 20, 21, 24]
plt.figure(figsize=(20, 8), dpi=80)
x_ticks_label = [f"{i}:00" for i in x]
plt.xticks(x, x_ticks_label, fontproperties=my_font, rotation=45)
y_ticks_label = [f'{i}' for i in range(15, 30)]
plt.yticks(range(15, 30), y_ticks_label)
plt.plot(x, y1, color='red', label='温度')
plt.plot(x, y2, color='blue', label='湿度')
plt.plot(x, y3, color='green', label='光照度')
plt.legend(prop=my_font, loc='upper left')
plt.show()
```

```python
#一图多个坐标系子图
import numpy as np
x=np.arange(1,100)
fig=plt.figure(figsize=(20,10), dpi=80)
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
ax1.plot(x,x)
ax1.grid(color='lightgray', linestyle='--',linewidth=0.5)
ax2.plot(x,-x)
ax3.plot(x,x**2)
ax4.plot(x,np.log(x))
plt.show()
```
